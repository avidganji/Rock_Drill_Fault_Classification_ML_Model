```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
import os
import seaborn as sns
sns.set_style('whitegrid')
# Set the visual style of seaborn plots to 'whitegrid' for better readability
```

```python
# Read the CSV file containing the feature data into a DataFrame
features = pd.read_csv('/Users//sarithavuppula//Downloads/Summer School/final_version_data.csv')
```

```python
# Filter the DataFrame to only include rows where the condition is 1
c1_features = features[features.condition==1]
c1_features_sample = c1_features.sample(n=1000, random_state=42)  # Adjust the sample size as needed

# Filter the DataFrame to only include rows where the mode is 2
m2_features = features[features['mode']==2]
m2_features_sample = m2_features.sample(n=1000, random_state=42)  # Adjust the sample size as needed
```

```python
# Create a pair plot to visualize the distribution of features by mode for condition 1
# This helps in understanding the relationships between different features for different fault modes
sns.pairplot(c1_features_sample.iloc[:,2:], hue='mode')
plt.show()
```

```python
# Create a pair plot to visualize the distribution of features by condition for mode 2
# This helps in understanding the relationships between different features for different conditions
sns.pairplot(m2_features_sample.drop(['cycle', 'mode'], axis=1), hue='condition')
plt.show()
```

```python
sns.set_style('whitegrid')
# Loop through each column in the DataFrame, starting from the 4th column onwards
for col in c1_features_sample.columns[3:]:
    plt.figure(figsize=(20,3))
    # Create a figure for each column with a specified size
    for m in c1_features['mode'].unique():
        # Loop through each unique mode in the 'mode' column
        m_c1_features = c1_features_sample[c1_features_sample['mode']==m]
        # Filter the DataFrame to only include rows where the mode matches the current mode in the loop
        plt.plot(m_c1_features[col])
        # Plot the values of the current column for the filtered DataFrame
        plt.title(col, fontsize=15)
        # Set the title of the plot to the current column name with a specified font size
```
